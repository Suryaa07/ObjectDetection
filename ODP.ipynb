```python
import cv2
import matplotlib.pyplot as plt
import os.path


fig, ax = plt.subplots()
ax.tick_params(labelbottom = "off", bottom = "off")
ax.tick_params(labelleft = "off", left = "off")
ax.set_xticklabels([])
ax.axis("off")


file = "cars1.jpg"
if os.path.exists(file):
  img = cv2.imread(file)
  show_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(show_img)
```
